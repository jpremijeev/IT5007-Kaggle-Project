# Data Preprocessing

Extract the prediction values

````python
y = data['Completion_rate'].values.reshape(-1,1)
y
````

Extraction features and check colinearity

````python
features = data.drop(['Unnamed: 0', 'Completion_rate'], axis=1)
features.corr()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(features.corr())
````

````python
plt.figure(figsize=(10,7))
sns.heatmap(features.corr()[np.abs(features.corr()>0.7)],annot=True,annot_kws={'size':15},cmap="YlGnBu")
plt.show()
````

### Creating a simple regression model

````python
from sklearn.linear_model import LinearRegression
features_selected = ['Tuition_out_state', 'Faculty_salary', 'SAT_average', 'Parents_college']
X = data[features_selected].values
model = LinearRegression()
model.fit(X, y)
score = model.score(X, y)
print(score)
````

````python
test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IT5006/Project/x_test.csv")
x_pred = test[features_selected]
print(x_pred)
y_pred = model.predict(x_pred)
print(y_pred)
````

### Creating our submission

````python
submission = pd.DataFrame(y_pred, columns = ['Completion_rate'])
submission
````

````python
submission.to_csv('submission.csv', index = True, index_label = "id") 
````
